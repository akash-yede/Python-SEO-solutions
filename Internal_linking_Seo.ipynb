<h1><b>Creating an internal linking visualization using Scrapy and Networkx</h1>

Interlinking is a process that where your web page has links to other pages on your website so that a user can click and be redirected to other pages that are related. Internal linking of a website has a significant impact on its ranking in search engines. It helps to reduce the website's bounce rate and increase average session duration by visitor.

To visualize how the web pages on our website are internally linked, we would require all the links on our website and graphically represent them. To fetch all the links, we'll use Scrapy, and further, to visualize, we'll use Networkx.

The second part involves using Networkx to visualize how these web pages are internally linked in the website. 

To create a visualization, we will create a graph to denote the web pages as nodes. We will use the edges as a connection between the web pages. 

To start with, we will start by importing networkx and create individual nodes and edges.

```
#Importing networkx
import networkx as nx

#Creating a graph object
G = nx.DiGraph()
```

```
#Creating variables to denote nodes

a = 'Home page'
b = 'Pricing page'
c = 'Enerprise page'
d = 'Our story page'
e = 'Blog page'
```

```
#Adding first node
G.add_node(a)

#Adding more nodes
G.add_nodes_from([b,c,d,e])

#Drawing the graph
nx.draw(G)
```

```
#Adding edges
G.add_edges_from([(a,b),(a,c),(a,d),(a,e),(b,a),(b,c),(b,d),(b,e),(c,a),(c,b),(c,d),(c,e),(d,a),(d,b),(d,c),(d,e),(e,a),(e,b),(e,c),(e,d)])

#Drawing the graph
nx.draw(G)
```

```
#Labeling the graph nodes
nx.draw_circular(G, with_labels=True)
```

The above graph represents that our website www.ranksense.com is strongly linked internally. All the five webpages that we took under consideration are internally linked.
